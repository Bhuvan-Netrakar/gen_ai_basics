```python
# Function to generate chatbot responses
def medical_chatbot(user_input, context="You are a helpful medical assistant."):
    prompt = f"{context}\nUser: {user_input}\nAI:"
    inputs = tokenizer(prompt, return_tensors="pt", padding=True)
    # The following two lines were previously in a separate cell/script
    # and need to be included within this function for input_ids and
    # attention_mask to be accessible to model.generate
    input_ids = inputs["input_ids"]
    attention_mask = inputs["attention_mask"]
    # Generate model output
    outputs = model.generate(
        input_ids=input_ids,
        attention_mask=attention_mask,
        max_length=200,
        temperature=0.7,
        top_p=0.9,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )
    # Decode and return the response
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response.split("AI:")[-1].strip()
```

```python
# Chat loop
print("Medical Chatbot (type 'exit' to quit)")
while True:
    user_input = input("You: ")
    if user_input.lower() in ['exit', 'quit']:
        break
    reply = medical_chatbot(user_input)
    print("Bot:", reply)
```
